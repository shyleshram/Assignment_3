```python
from collections import Counter
import csv
```

```python
with open('data.csv') as csv_file:
    csv_pointer = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_pointer:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
  

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print(popularity)

# print(language_counter.most_common(15))

```

```python
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.title('Most popular Languages')
plt.xlabel('No of people used')
plt.ylabel('Programming Languages')
```

```python
data1 = pd.read_csv('data.csv')
ids = data1['Responder_id']
lang_responses = data1['LanguagesWorkedWith']

lang_counter = Counter()

for response in lang_responses:
    lang_counter.update(response.split(';'))
    
L = []
P = []

for item in lang_counter.most_common(15):
    L.append(item[0])
    P.append(item[1])

L.reverse()
P.reverse()
print(L)
print(P)

plt.barh(L,P)
```
