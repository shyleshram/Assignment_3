```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('data2.csv')
```

```python
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages,dev_salaries,'--k',label='All Devs')
plt.plot(ages,py_salaries,label='Python')
plt.legend()
```

```python
plt.plot(ages,dev_salaries,'--k',label='All Devs')
plt.plot(ages,py_salaries,label='Python')

# fill under python plotted data
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25,
                where=(py_salaries > overall_median),
                interpolate=True)
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25,
                where=(py_salaries <= overall_median), color='red',
                interpolate=True)

plt.legend()
plt.tight_layout()
```

```python
plt.plot(ages,dev_salaries,'--k',label='All Devs')
plt.plot(ages,py_salaries,label='Python')

# fill under python plotted data

plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.25,
                where=(py_salaries > dev_salaries),
                interpolate=True, label='Above_Average')
plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.25,
                where=(py_salaries <= dev_salaries), color='red',
                interpolate=True, label='Below_Average')

plt.legend()
plt.tight_layout()
```
